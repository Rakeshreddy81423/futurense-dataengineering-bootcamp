```python
sectorGrp = df.groupby('Sector')

#count of groups formed
len(sectorGrp)
```

```python


sectorGrp[['Revenue','Profits']].agg({'Revenue':'sum','Profits':'max'})
```

```python
def findMiddleChars(string,N):
    mid = len(string) //2 
    half = N //2
    start = mid - half
    end = mid + half
    if N % 2 == 1:
        print(string[start:end+1])
    else:
        print(string[start:end])
    

str1 = "JohnDipTest"

findMiddleChars(str1,4)
```

```python
# groups and rows in each group
sectorGrp.groups
```
