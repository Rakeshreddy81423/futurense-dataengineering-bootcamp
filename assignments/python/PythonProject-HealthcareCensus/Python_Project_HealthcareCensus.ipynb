## Problem Statement 3: (Rename State/UT Names)

#### The State/UT names are in all caps in the census data, For uniformity across datasets we use the names so that only the first character of each word in the name is in upper case and the rest are in lower case. However, if the word is “and” then it should be  all lowercase.

```python
def change_state_upper_to_lower(state_ut):
    
    new_state_ut = []
    
    for word in state_ut.split():
        if word != "AND":
            new_state_ut.append(str(word[0]+word[1:].lower()))
        else:
            new_state_ut.append(word.lower())
        
    return " ".join(new_state_ut)


census_df["State/UT"] = census_df["State/UT"].apply(lambda state_ut : change_state_upper_to_lower(state_ut))

census_df
            
    
```

# Problem Statement 5: (Find and process Missing Data)

### Fill population column
#### Population = Male + Female  (or)
#### Population  = Young_and_Adult+  Middle_Aged + Senior_Citizen + Age_Not_Stated

```python
'''
function takes row as input

if population value is nan then:
        check if any of male or female is nan then :
            find total population by adding different age grps population
        else return sum of male and female
else return the population value
        
'''
def fill_population(row):
    if np.isnan(row['Population']):
        if np.isnan(row['Male']) or np.isnan(row['Female']):
             pop = row['Young_and_Adult'] + row['Middle_Aged'] + row['Senior_Citizen'] + row['Age_Not_Stated']
        else:
             pop = row['Male'] + row ['Female']
    else:
        pop = row['Population']
        
    return pop
        
            
            

census_df['Population'] = census_df.apply(lambda row: fill_population(row),axis=1)

```

```python
census_df[census_df['Population'].isna()]
```

### Fill Literate column
#### Literate = Literate_Male + Literate_Female

```python
def fill_literate(row):
    if np.isnan(row['Literate']):
        return row['Literate_Male'] + row['Literate_Female']
    return row['Literate']
    
    
    
census_df['Literate'] = census_df.apply(lambda row: fill_literate(row),axis=1)
```

```python
census_df[census_df['Literate'].isna()]
```

### Fill Households
#### Households   =   Households_Rural + Households_Urban

```python
def fill_households(row):
    if np.isnan(row['Households']):
        return row['Households_Rural'] + row['Households_Urban']
    return row['Households']
    

census_df['Households'] = census_df.apply(lambda row: fill_households(row),axis=1)
```

```python
census_df[census_df['Households'].isna()]
```

### Fill Male and Female Column
#### Male = Population - Female 
##### Female = Population - Male

```python
def fill_Male(row):
    if np.isnan(row['Male']):
        if np.isnan(row['Population']) or np.isnan(row['Female']):
            return row['Male']
        else:
            return row['Population'] - row['Female']
    else:
        return row['Male']
        
def fill_Female(row):
    if np.isnan(row['Female']):
        if np.isnan(row['Population']) or np.isnan(row['Male']):
            return row['Female']
        else:
            return row['Population'] - row['Male']
    else:
        return row['Female']
    
    
census_df['Female'] = census_df.apply(lambda row: fill_Female(row),axis=1) 
census_df['Male'] = census_df.apply(lambda row: fill_Male(row),axis=1)  
```

```python
census_df[census_df['Male'].isna()]
```

```python
census_df[census_df['Female'].isna()]
```

### Fill Literate_Male and Literate_Female
#### Literate_Male = Literate - Literate_Female
#### Literate_Female = Literate - Literate_Male

```python

def fill_literate_male(row):
    if np.isnan(row['Literate_Male']):
        if np.isnan(row['Literate']) or np.isnan(row['Literate_Female']):
            return row['Literate_Male']
        else:
            return row['Literate'] - row['Literate_Female']
    else:
        return row['Literate_Male']
    
def fill_literate_female(row):
    if np.isnan(row['Literate_Female']):
        if np.isnan(row['Literate']) or np.isnan(row['Literate_Male']):
            return row['Literate_Female']
        else:
            return row['Literate'] - row['Literate_Male']
    else:
        return row['Literate_Female']
    
    
census_df['Literate_Male'] = census_df.apply(lambda row : fill_literate_male(row),axis=1)
census_df['Literate_Female'] = census_df.apply(lambda row : fill_literate_female(row),axis=1)
```

```python
census_df[census_df['Literate_Male'].isna()]
```

```python
census_df[census_df['Literate_Female'].isna()]
```

### Fill Households_Rural and Households_Urban
#### Households_Rural = Households - Households_Urban
#### Households_Urban = Households - Households_Rural

```python
def fill_households_rural(row):
    if np.isnan(row['Households_Rural']):
        if np.isnan(row['Households']) or np.isnan(row['Households_Urban']):
            return row['Households_Rural']
        else:
            return row['Households'] - row['Households_Urban']
    else:
        return row['Households_Rural']
    
def fill_households_urban(row):
    if np.isnan(row['Households_Urban']):
        if np.isnan(row['Households']) or np.isnan(row['Households_Rural']):
            return row['Households_Urban']
        else:
            return row['Households'] - row['Households_Rural']
    else:
        return row['Households_Urban']
    
    
census_df['Households_Rural'] = census_df.apply(lambda row : fill_households_rural(row),axis=1)
census_df['Households_Urban'] = census_df.apply(lambda row : fill_households_urban(row),axis=1)
            
```

```python
census_df[census_df['Households_Rural'].isna()]
```

```python
census_df[census_df['Households_Urban'].isna()]
```

### Fill different age_groups
#### age_groups = population - (remaining age_groups)

```python
def fill_age_group_columns(row,group):
    li = ['Young_and_Adult','Middle_Aged','Senior_Citizen','Age_Not_Stated']
    li.remove(group)
    
    if np.isnan(row[group]):
        if np.isnan(row['Population']) or np.isnan(row[li[0]]) or np.isnan(row[li[1]]) or np.isnan(row[li[2]]):
            return row[group]
        else:
            return row['Population'] - (row[li[0]] + row[li[1]] + row[li[2]])
    else:
        return row[group]
    
    
census_df['Young_and_Adult'] = census_df.apply(lambda row : fill_age_group_columns(row,'Young_and_Adult'),axis=1)
census_df['Middle_Aged'] = census_df.apply(lambda row : fill_age_group_columns(row,'Middle_Aged'),axis=1)
census_df['Senior_Citizen'] = census_df.apply(lambda row : fill_age_group_columns(row,'Senior_Citizen'),axis=1)
census_df['Age_Not_Stated'] = census_df.apply(lambda row : fill_age_group_columns(row,'Age_Not_Stated'),axis=1)


    
```

```python
current_percentages = census_df.isna().sum() * 100 / len(census_df)
```

```python
df = pd.concat([prev_percentages,current_percentages],keys=["percentage_before_cleaning","percentage_after_cleaning"],axis=1)

fig = df.plot(kind='bar',title = "Comparison of NaN percentages across all the columns before and after cleaning the data")

fig.set(xlabel="Columns",ylabel="Percentage")

fig;
```
